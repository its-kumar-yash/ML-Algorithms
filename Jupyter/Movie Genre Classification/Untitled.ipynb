# 10. In Which Year There Was The Highest Voting?

```python
df.groupby('Year')['Votes'].mean().sort_values(ascending = False)
```

```python
sns.barplot(x='Year', y='Votes', data=df)
plt.title("Votes By Year")
```

# 13. Display Top 10 Lengthy Movies Title

```python
df.columns
```

```python
df1 = df.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']].set_index('Title')
```

```python
sns.barplot(x='Runtime (Minutes)',y=df1.index, data=df1)
plt.title('Top 5 Lengtly Movies')
plt.show()
```
