# Relational/Statistical Plots

<ol>
    <li>Scatter Plots</li>

## Scatter Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
# Plot Scatter Plot b/w total_bill and tip
sns.relplot(x='total_bill', y='tip', kind='scatter', data=data)
```

```python
# hue Parameter
sns.relplot(x='total_bill', y='tip',hue='smoker', data=data)
```

```python
#style parameter
sns.relplot(x='total_bill', y='tip',hue='smoker', style='sex', data=data)
```

```python
#size parameter
sns.relplot(x='total_bill', y='tip',hue='smoker', style='sex', size='size', data=data)
```

```python
# Using scatterplot function
sns.scatterplot(x='total_bill', y='tip',hue='smoker', data=data)
```

# Plotting with Categorical Data

## catplot [Figure Level Function]

### 1. Categorical Scatterplots
##### a. Stripplot
##### b. Swarmplot 

### 2. Categorical Distribution Plot
##### a. Boxplot
##### b. ViolinPlot

### 3. Categorical estimate Plot

##### c. Countplot

### 4. Categorical Categorical Plot
##### a. Heatmaps

### 5. Plotting side by side graphs using facetGrids

```python
# Categorical ScatterPlot
```

```python
# x-axis - Categorical Data
# y-axis - Numerical Data
```

```python
sns.catplot(x='day', y='tip', kind='strip', data=data)
#Jitter
```

```python
sns.catplot(x='day', y='tip', kind='swarm', data=data)
```

```python
sns.catplot(x='day', y='tip', kind='strip', hue='sex', data=data)
```

```python
sns.swarmplot(x='day', y='tip', data=data)
```

```python
# BoxPlot
```

```python
# univariate analysis
sns.boxplot(data['total_bill'])
```

```python
# multivariate
sns.catplot(x='day', y='total_bill', kind='box', data=data)
```

```python
sns.catplot(x='day', y='total_bill', kind='box', hue='sex', data=data)
```

```python
# Violin Plot
sns.violinplot(data['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=data)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', hue='sex', split=True, data=data)
```

```python
# Barplot and CountPlot
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', data=data)
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', data=data)
```

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', hue='sex', estimator=np.std , data=data)
```

```python
# countplot
sns.catplot(x='sex', kind='count',data=data)
```

```python
sns.catplot(x='sex', kind='count', hue = 'smoker', data=data)
```

```python
# HeatMap
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x = flights.pivot_table(index='year', columns='month',  values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidth=0.2, annot=True, fmt='d')
```

```python
sns.heatmap(x, cbar=False, linewidth=0.2, annot=True, fmt='d', cmap='viridis')
```

```python
sns.heatmap(x, cbar=False, linewidth=0.2, annot=True, fmt='d', cmap='summer')
```

```python
sns.heatmap(x, cbar=False, linewidth=0.2, annot=True, fmt='d', cmap='winter')
```

```python
#Cluster Map
sns.clustermap(x, z_score=0,annot=True, metric='correlation')
```

```python
# Joint Plot
```

```python
sns.jointplot(x='total_bill', y='tip', data=data)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=data)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde', data=data)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=data)
```
