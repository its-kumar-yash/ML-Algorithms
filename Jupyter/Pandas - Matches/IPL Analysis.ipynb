```python
# find out top 5 batsmen in term of run scored
delivery = pd.read_csv('deliveries.csv')
delivery.head()
```

```python
#against which team vk scored most runs
vk = delivery[delivery['batsman'] == 'V Kohli']
```

```python
mostruns = vk.groupby('bowling_team')
```

```python
mostruns['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
def most_runs(batsman_name):
    vk = delivery[delivery['batsman'] == batsman_name]
    mostruns = vk.groupby('bowling_team')
    return mostruns['batsman_runs'].sum().sort_values(ascending=False).head(3)
```

```python
most_runs('DA Warner')
```

```python
# isin()
#find the most destructive death over batsmen in the history of IPL in last 5 overs
# on the basis of strike rate
# strike rate = (num of runs/num of balls)/100
# min batsman 200 balls in overs 16-20
```

```python
#extract death over
death_over = delivery[delivery['over']>15]
```

```python
#balls played by batmens in death overs
all_batsmen = death_over.groupby('batsman')['batsman_runs'].count()
x = all_batsmen > 200
batsmen_list = all_batsmen[x].index.tolist()
```

```python
#fina strike rate of this batsmen list
#runs scored by these batsmen and balls played by these batsmen
final = delivery[delivery['batsman'].isin(batsmen_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```

```python
sr
```
